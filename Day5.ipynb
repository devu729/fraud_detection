```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('data/creditcard.csv')
```

```python
print(df['Class'].value_counts())
```

```python
fraud_counts = df['Class'].sum()
non_fraud_counts = df['Class'].count() - fraud_counts
fraud_ratio = fraud_counts / df['Class'].count()

print(f"Fraudulant transaction: {fraud_counts}")
print(f"Non- Fraudulant transaction: {non_fraud_counts}")
print(f"Fraudulant transaction rate: {fraud_ratio}")
```

```python
plt.figure(figsize=(6,8))
sns.countplot(x='Class', data = df)
plt.xticks([0,1], ['Non-Fraud', 'Fraud'])
plt.title("Class distribution")
plt.show()
```

```python
plt.figure(figsize=(6,8))
labels = ['non-fraud', 'fraud']
sizes = [fraud_counts, non_fraud_counts]
plt.pie(sizes, labels=labels, autopct="%1.5f%%", startangle=90)
plt.show()
```
